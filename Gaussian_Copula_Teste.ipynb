```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
================================================================================
GAUSSIAN COPULA + BOOTSTRAP PARA DATASET TIPS (Restaurant Tips)
Versão com 2 contínuas e 5 ordinais (size, day, time, smoker, sex)
================================================================================
"""

# ============================================================================
# BLOCO 1: IMPORTAÇÕES E CONFIGURAÇÕES INICIAIS
# ============================================================================

import subprocess
import sys
import warnings
import time
import random
from IPython.display import display, HTML

CSS_STYLES = """
<style>
    .header {
        background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        color: white;
        padding: 15px;
        border-radius: 10px;
        margin: 10px 0;
        text-align: center;
        font-family: Arial, sans-serif;
        font-weight: bold;
        font-size: 18px;
    }
    .success-box {
        background-color: #d4edda;
        color: #155724;
        padding: 12px;
        border: 1px solid #c3e6cb;
        border-radius: 5px;
        margin: 10px 0;
        font-weight: bold;
    }
    .warning-box {
        background-color: #fff3cd;
        color: #856404;
        padding: 12px;
        border: 1px solid #ffeaa7;
        border-radius: 5px;
        margin: 10px 0;
        font-weight: bold;
    }
    .info-box {
        background-color: #d1ecf1;
        color: #0c5460;
        padding: 12px;
        border: 1px solid #bee5eb;
        border-radius: 5px;
        margin: 10px 0;
    }
    .metrics-highlight {
        background: linear-gradient(135deg, #ffd89b 0%, #19547b 100%);
        color: white;
        padding: 15px;
        border-radius: 10px;
        margin: 10px 0;
        text-align: center;
        font-weight: bold;
    }
    .dataframe-table {
        font-family: Arial, sans-serif;
        border-collapse: collapse;
        width: 100%;
        box-shadow: 0 4px 8px rgba(0,0,0,0.1);
        border-radius: 10px;
        overflow: hidden;
        margin: 20px 0;
    }
    .dataframe-table th {
        background: linear-gradient(135deg, #2c3e50 0%, #3498db 100%);
        color: white;
        padding: 12px;
        text-align: left;
        font-weight: bold;
    }
    .dataframe-table td {
        padding: 10px;
        border-bottom: 1px solid #ddd;
    }
    .dataframe-table tr:nth-child(even) {
        background-color: #f8f9fa;
    }
    .dataframe-table tr:hover {
        background-color: #e9ecef;
        transform: scale(1.01);
        transition: all 0.2s ease;
    }
    .rubin-rules-box {
        background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        color: white;
        padding: 15px;
        border-radius: 10px;
        margin: 10px 0;
        text-align: center;
        font-weight: bold;
    }
    .iteration-box {
        background-color: #e8f4fd;
        border-left: 5px solid #3498db;
        padding: 10px;
        margin: 10px 0;
        border-radius: 5px;
    }
    .encoding-badge {
        background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%);
        color: white;
        padding: 5px 10px;
        border-radius: 15px;
        font-size: 12px;
        font-weight: bold;
        display: inline-block;
        margin: 2px;
    }
</style>
"""
display(HTML(CSS_STYLES))

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

for lib in ['numpy', 'pandas', 'scipy', 'scikit-learn', 'matplotlib', 'seaborn']:
    try:
        __import__(lib)
    except ImportError:
        install(lib)

import numpy as np
import pandas as pd
from scipy import stats
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, cohen_kappa_score, matthews_corrcoef)
import matplotlib.pyplot as plt
import seaborn as sns

try:
    from gcimpute.gaussian_copula import GaussianCopula
except ImportError:
    print("Instalando gcimpute do GitHub...")
    subprocess.check_call([sys.executable, "-m", "pip", "install",
                           "git+https://github.com/udellgroup/gcimpute.git"])
    from gcimpute.gaussian_copula import GaussianCopula

warnings.filterwarnings("ignore")
if not hasattr(np, 'round_'):
    np.round_ = np.round

# ============================================================================
# BLOCO 2: CONFIGURAÇÕES GLOBAIS
# ============================================================================

RANDOM_SEED = 42
BOOTSTRAP_ITERACOES = 100
M_IMPUTATIONS = 20
ALPHA = 0.05
MAX_ITER_EM = 200
TOL_EM = 0.001
np.random.seed(RANDOM_SEED)

# ============================================================================
# BLOCO 3: CARREGAMENTO E PRÉ-PROCESSAMENTO (TIPS) - 2 CONTÍNUAS, 5 ORDINAIS
# ============================================================================

display(HTML('<div class="header">🍽️ CARREGANDO DATASET TIPS (2 contínuas, 5 ordinais)</div>'))

tips = sns.load_dataset("tips")
df_raw = tips.copy()
display(HTML(f'<div class="success-box">✅ Dados carregados: {df_raw.shape[0]} linhas, {df_raw.shape[1]} colunas.</div>'))
display(HTML('<div class="info-box"><b>Primeiras linhas:</b></div>'))
display(df_raw.head().style.set_table_attributes('class="dataframe-table"'))

df_clean = df_raw.copy()

# Definição dos tipos
col_continuas = ["total_bill", "tip"]
col_ordinais = ["size", "day", "time", "smoker", "sex"]
col_nominais = []

# Define ordem para as categóricas (exceto size, que já é int)
df_clean['day'] = pd.Categorical(df_clean['day'], categories=['Thur','Fri','Sat','Sun'], ordered=True)
df_clean['time'] = pd.Categorical(df_clean['time'], categories=['Lunch','Dinner'], ordered=True)
df_clean['smoker'] = pd.Categorical(df_clean['smoker'], categories=['No','Yes'], ordered=True)
df_clean['sex'] = pd.Categorical(df_clean['sex'], categories=['Female','Male'], ordered=True)
# size permanece int

niveis_originais = {}
for col in col_ordinais + col_nominais:
    df_clean[col] = df_clean[col].astype("category")
    niveis_originais[col] = list(df_clean[col].cat.categories)

display(HTML('<div class="info-box"><b>Níveis das variáveis ordinais (com ordem):</b></div>'))
for col, lvls in niveis_originais.items():
    display(HTML(f'<span class="encoding-badge">{col}: {lvls}</span> '))

# Codificação para inteiros (preserva ordem)
df_cod = df_clean.copy()
for col in col_ordinais + col_nominais:
    df_cod[col] = df_cod[col].cat.codes

display(HTML('<div class="info-box"><b>Dados codificados (head):</b></div>'))
display(df_cod.head().style.set_table_attributes('class="dataframe-table"'))

colunas_ordenadas = list(df_cod.columns)
indices_continuas = [colunas_ordenadas.index(col) for col in col_continuas]
indices_ordinais = [colunas_ordenadas.index(col) for col in (col_ordinais + col_nominais)]

print("Índices contínuas:", indices_continuas)
print("Índices ordinais :", indices_ordinais)

# ============================================================================
# BLOCO 4: FUNÇÕES AUXILIARES (MISSING, IMPUTAÇÃO, MÉTRICAS)
# ============================================================================

def criar_missing_mcar_estratificado(df, percentual=0.75, random_seed=42):
    np.random.seed(random_seed)
    random.seed(random_seed)
    df_com_na = df.copy()
    posicoes_originais = {}
    for col in df.columns:
        n_missing = int(len(df) * percentual)
        indices = np.random.choice(df.index, size=n_missing, replace=False)
        for i in indices:
            posicoes_originais[(i, col)] = df.loc[i, col]
            df_com_na.loc[i, col] = np.nan
    return df_com_na, posicoes_originais

def imputacoes_gcimpute(df_na, indices_continuas, indices_ordinais,
                        m=M_IMPUTATIONS, random_seed=None, verbose=True,
                        maxit=MAX_ITER_EM, tol=TOL_EM,
                        use_conditional_mean=False):
    X = df_na.astype(float).values
    model = GaussianCopula(
        max_iter=maxit, tol=tol,
        random_state=random_seed if random_seed is not None else RANDOM_SEED,
        verbose=1 if verbose else 0,
    )
    if verbose:
        print(f"  Ajustando Gaussian Copula (EM)...")
        print(f"    contínuas: {indices_continuas}")
        print(f"    ordinais : {indices_ordinais}")
    model.fit(X, continuous=indices_continuas, ordinal=indices_ordinais)
    if use_conditional_mean:
        X_imp = model.transform(X)
        df_imp = pd.DataFrame(X_imp, columns=df_na.columns, index=df_na.index)
        datasets = [df_imp]
    else:
        X_imp = model.sample_imputation(X=X, num=m)
        datasets = [pd.DataFrame(X_imp[:, :, i], columns=df_na.columns, index=df_na.index) for i in range(m)]
    return datasets

def metricas_continuas(reais, imputados):
    mask = ~np.isnan(reais) & ~np.isnan(imputados)
    if mask.sum() < 2:
        return dict(rmse=np.nan, mae=np.nan, mse=np.nan, cor=np.nan, nrmse=np.nan,
                    nmae=np.nan, mape=np.nan, smape=np.nan, r2=np.nan)
    r = reais[mask]
    p = imputados[mask]
    mse = np.mean((r - p) ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(r - p))
    cor = np.corrcoef(r, p)[0, 1] if len(r) > 1 else np.nan
    intervalo = np.ptp(r)
    nrmse = rmse / intervalo if intervalo > 0 else np.nan
    nmae = mae / intervalo if intervalo > 0 else np.nan
    with np.errstate(divide='ignore', invalid='ignore'):
        mape = np.mean(np.abs((r - p) / (r + 1e-8))) * 100
        mape = mape if np.isfinite(mape) else np.nan
    with np.errstate(divide='ignore', invalid='ignore'):
        denom = (np.abs(r) + np.abs(p) + 1e-8)
        smape = np.mean(2 * np.abs(r - p) / denom) * 100
        smape = smape if np.isfinite(smape) else np.nan
    ss_res = np.sum((r - p) ** 2)
    ss_tot = np.sum((r - np.mean(r)) ** 2)
    r2 = 1 - (ss_res / ss_tot) if ss_tot > 0 else np.nan
    return dict(rmse=rmse, mae=mae, mse=mse, cor=cor, nrmse=nrmse,
                nmae=nmae, mape=mape, smape=smape, r2=r2)

def erro_matriz_correlacao(df_original, df_imputado, col_continuas):
    orig = df_original[col_continuas].dropna()
    imp = df_imputado.loc[orig.index, col_continuas]
    cor_orig = orig.corr().values
    cor_imp = imp.corr().values
    diff = np.linalg.norm(cor_imp - cor_orig, 'fro')
    norm_orig = np.linalg.norm(cor_orig, 'fro')
    return diff / norm_orig if norm_orig > 0 else np.nan

def calcular_smae_por_coluna(valores_reais, valores_imputados, referencia):
    if len(valores_reais) == 0:
        return np.nan
    reais = np.array(valores_reais)
    imputados = np.array(valores_imputados)
    erro_imp = np.sum(np.abs(imputados - reais))
    erro_ref = np.sum(np.abs(referencia - reais))
    if erro_ref == 0:
        return np.nan
    return erro_imp / erro_ref

def gower_distance_matrix(df_real, df_imputado, col_continuas, col_categoricas):
    n = len(df_real)
    if n == 0:
        return np.nan
    all_cols = col_continuas + col_categoricas
    df_real_sub = df_real[all_cols].reset_index(drop=True)
    df_imp_sub = df_imputado[all_cols].reset_index(drop=True)
    gower_distances = np.zeros(n)
    for col in col_continuas:
        real_vals = df_real_sub[col].values
        imp_vals = df_imp_sub[col].values
        rng = np.ptp(real_vals)
        if rng == 0:
            rng = 1.0
        diff = np.abs(real_vals - imp_vals) / rng
        gower_distances += diff
    for col in col_categoricas:
        real_vals = df_real_sub[col].values
        imp_vals = df_imp_sub[col].values
        diff = (real_vals != imp_vals).astype(float)
        gower_distances += diff
    n_cols = len(all_cols)
    gower_distances /= n_cols
    return np.mean(gower_distances)

# ============================================================================
# BLOCO 5: REGRAS DE RUBIN
# ============================================================================

def rubin_pool(estimativas, m, alpha=0.05):
    estimativas = np.array(estimativas, dtype=float)
    estimativas = estimativas[~np.isnan(estimativas)]
    k = len(estimativas)
    if k == 0:
        return {'Q_bar': np.nan, 'se': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
    Q_bar = np.mean(estimativas)
    if k == 1:
        return {'Q_bar': Q_bar, 'se': 0.0, 'ci_lower': Q_bar, 'ci_upper': Q_bar}
    B = np.var(estimativas, ddof=1)
    T = (1 + 1/m) * B
    se = np.sqrt(T)
    gl = m - 1
    t_val = stats.t.ppf(1 - alpha/2, gl)
    ci_lower = Q_bar - t_val * se
    ci_upper = Q_bar + t_val * se
    return {'Q_bar': Q_bar, 'se': se, 'ci_lower': ci_lower, 'ci_upper': ci_upper}

# ============================================================================
# BLOCO 6: FUNÇÃO PRINCIPAL DE IMPUTAÇÃO + AVALIAÇÃO
# ============================================================================

def multiple_imputation_rubin_misto(df_na, posicoes_originais, df_original_clean,
                                    col_continuas, col_ordinais, col_nominais,
                                    indices_continuas, indices_ordinais,
                                    m=M_IMPUTATIONS, random_seed=RANDOM_SEED, show_progress=True,
                                    use_conditional_mean=False):
    # SMAE sobre contínuas e ordinais (todas as ordinais)
    col_continuas_smae = col_continuas                     # total_bill, tip
    col_ordinais_smae = col_ordinais                       # size, day, time, smoker, sex

    if show_progress:
        if use_conditional_mean:
            display(HTML(f'<div class="iteration-box">🔹 Gerando imputação por média condicional...</div>'))
        else:
            display(HTML(f'<div class="iteration-box">🔹 Gerando {m} imputações via gcimpute...</div>'))
    datasets = imputacoes_gcimpute(df_na, indices_continuas, indices_ordinais,
                                   m=m, random_seed=random_seed, verbose=show_progress,
                                   use_conditional_mean=use_conditional_mean)
    n_imp = len(datasets)
    if n_imp == 0:
        return dict(datasets=[], metricas_cat={}, metricas_cont={})

    medianas_cont = {col: df_original_clean[col].median() for col in col_continuas_smae}
    modas_cat = {}
    for col in col_ordinais_smae:
        moda = df_original_clean[col].mode()
        modas_cat[col] = moda.iloc[0] if len(moda) > 0 else np.nan

    acc_list, prec_list, rec_list, f1_list, kappa_list, mcc_list = [], [], [], [], [], []
    erro_cor_list = []
    dict_cont = {col: {met: [] for met in ['rmse','mae','mse','cor','nrmse','nmae','mape','smape','r2']} for col in col_continuas}
    smae_cont_por_coluna = {col: [] for col in col_continuas_smae}
    smae_cat_por_coluna = {col: [] for col in col_ordinais_smae}
    cols_categoricas = col_ordinais + col_nominais   # todas as categóricas (aqui são apenas ordinais)
    gower_list = []

    for i, df_imp in enumerate(datasets):
        if show_progress:
            display(HTML(f'<div class="metrics-highlight">📊 Avaliando imputação {i+1}/{n_imp}</div>'))
        err_cor = erro_matriz_correlacao(df_original_clean, df_imp, col_continuas)
        erro_cor_list.append(err_cor)

        # Avaliação categórica (todas as ordinais)
        reais_cat, preditos_cat = [], []
        reais_cat_por_col = {col: [] for col in cols_categoricas}
        imps_cat_por_col = {col: [] for col in cols_categoricas}
        for (linha, coluna), valor_real in posicoes_originais.items():
            if coluna in cols_categoricas:
                valor_imp = df_imp.at[linha, coluna]
                if pd.notna(valor_real) and pd.notna(valor_imp):
                    pred = int(round(valor_imp))
                    real = int(valor_real)
                    k = len(niveis_originais.get(coluna, [0]))
                    pred = max(0, min(pred, k-1))
                    reais_cat.append(real)
                    preditos_cat.append(pred)
                    reais_cat_por_col[coluna].append(real)
                    imps_cat_por_col[coluna].append(pred)
        if reais_cat:
            try:
                acc_list.append(accuracy_score(reais_cat, preditos_cat))
                prec_list.append(precision_score(reais_cat, preditos_cat, average='macro', zero_division=0))
                rec_list.append(recall_score(reais_cat, preditos_cat, average='macro', zero_division=0))
                f1_list.append(f1_score(reais_cat, preditos_cat, average='macro', zero_division=0))
                kappa_list.append(cohen_kappa_score(reais_cat, preditos_cat))
                mcc_list.append(matthews_corrcoef(reais_cat, preditos_cat))
            except:
                pass
            # SMAE ordinal (para cada coluna ordinal)
            for col in col_ordinais_smae:
                if col in reais_cat_por_col and reais_cat_por_col[col]:
                    smae = calcular_smae_por_coluna(reais_cat_por_col[col], imps_cat_por_col[col], modas_cat[col])
                    smae_cat_por_coluna[col].append(smae)

        # Avaliação contínua
        for col in col_continuas:
            reais_cont, imps_cont = [], []
            for (linha, coluna), valor_real in posicoes_originais.items():
                if coluna == col:
                    imp_val = df_imp.at[linha, col]
                    if pd.notna(valor_real) and pd.notna(imp_val):
                        reais_cont.append(valor_real)
                        imps_cont.append(imp_val)
            if reais_cont:
                met = metricas_continuas(np.array(reais_cont), np.array(imps_cont))
                for mname in dict_cont[col]:
                    dict_cont[col][mname].append(met[mname])
                if col in col_continuas_smae:
                    smae = calcular_smae_por_coluna(reais_cont, imps_cont, medianas_cont[col])
                    smae_cont_por_coluna[col].append(smae)

        # Distância de Gower
        real_dict = {}
        imp_dict = {}
        for (linha, coluna), val_real in posicoes_originais.items():
            if linha not in real_dict:
                real_dict[linha] = {}
                imp_dict[linha] = {}
            real_dict[linha][coluna] = val_real
            imp_dict[linha][coluna] = df_imp.at[linha, coluna]
        real_rows, imp_rows = [], []
        for linha in real_dict.keys():
            row_real = {}
            row_imp = {}
            for col in col_continuas + cols_categoricas:
                if col in real_dict[linha]:
                    row_real[col] = real_dict[linha][col]
                    row_imp[col] = imp_dict[linha][col]
                else:
                    row_real[col] = df_original_clean.at[linha, col]
                    row_imp[col] = df_original_clean.at[linha, col]
            real_rows.append(row_real)
            imp_rows.append(row_imp)
        df_real_gower = pd.DataFrame(real_rows)
        df_imp_gower = pd.DataFrame(imp_rows)
        gower_list.append(gower_distance_matrix(df_real_gower, df_imp_gower, col_continuas, cols_categoricas))

    # Agregação
    if use_conditional_mean:
        metricas_cat = {
            'acuracia': acc_list[0] if acc_list else np.nan,
            'precisao': prec_list[0] if prec_list else np.nan,
            'recall': rec_list[0] if rec_list else np.nan,
            'f1_score': f1_list[0] if f1_list else np.nan,
            'kappa': kappa_list[0] if kappa_list else np.nan,
            'mcc': mcc_list[0] if mcc_list else np.nan,
            'erro_correlacao': erro_cor_list[0] if erro_cor_list else np.nan,
            'gower_distance': gower_list[0] if gower_list else np.nan
        }
        smae_cat_col_means = {col: np.mean(smae_cat_por_coluna.get(col, [])) if smae_cat_por_coluna.get(col) else np.nan for col in col_ordinais_smae}
        metricas_cat['smae_medio'] = np.nanmean(list(smae_cat_col_means.values()))

        metricas_cont = {}
        for col in col_continuas:
            metricas_cont[col] = {met: dict_cont[col][met][0] if dict_cont[col][met] else np.nan for met in dict_cont[col]}
        smae_cont_col_means = {col: np.mean(smae_cont_por_coluna.get(col, [])) if smae_cont_por_coluna.get(col) else np.nan for col in col_continuas_smae}
        metricas_cont['smae_medio'] = np.nanmean(list(smae_cont_col_means.values()))
        # Agregados
        metricas_cont['nrmse_medio'] = np.mean([dict_cont[col]['nrmse'][0] for col in col_continuas if dict_cont[col]['nrmse']]) if any(dict_cont[col]['nrmse'] for col in col_continuas) else np.nan
        metricas_cont['nmae_medio'] = np.mean([dict_cont[col]['nmae'][0] for col in col_continuas if dict_cont[col]['nmae']]) if any(dict_cont[col]['nmae'] for col in col_continuas) else np.nan
        metricas_cont['mape_medio'] = np.mean([dict_cont[col]['mape'][0] for col in col_continuas if dict_cont[col]['mape']]) if any(dict_cont[col]['mape'] for col in col_continuas) else np.nan
        metricas_cont['smape_medio'] = np.mean([dict_cont[col]['smape'][0] for col in col_continuas if dict_cont[col]['smape']]) if any(dict_cont[col]['smape'] for col in col_continuas) else np.nan
        metricas_cont['r2_medio'] = np.mean([dict_cont[col]['r2'][0] for col in col_continuas if dict_cont[col]['r2']]) if any(dict_cont[col]['r2'] for col in col_continuas) else np.nan

        metricas_cat_rubin = {k: {'Q_bar': v, 'ci_lower': np.nan, 'ci_upper': np.nan} for k, v in metricas_cat.items()}
        metricas_cont_rubin = {}
        for k, v in metricas_cont.items():
            if isinstance(v, dict):
                metricas_cont_rubin[k] = {kk: vv for kk, vv in v.items()}
            else:
                metricas_cont_rubin[k] = {'Q_bar': v, 'ci_lower': np.nan, 'ci_upper': np.nan}
        return dict(datasets=datasets, metricas_cat=metricas_cat_rubin, metricas_cont=metricas_cont_rubin)
    else:
        # Múltiplas imputações com Rubin
        metricas_cat = {}
        for nome, lst in [('acuracia', acc_list), ('precisao', prec_list), ('recall', rec_list),
                          ('f1_score', f1_list), ('kappa', kappa_list), ('mcc', mcc_list)]:
            metricas_cat[nome] = rubin_pool(lst, n_imp, ALPHA) if lst else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cat['erro_correlacao'] = rubin_pool(erro_cor_list, n_imp, ALPHA) if erro_cor_list else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cat['gower_distance'] = rubin_pool(gower_list, n_imp, ALPHA) if gower_list else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        smae_cat_col_means = {col: np.mean(smae_cat_por_coluna.get(col, [])) if smae_cat_por_coluna.get(col) else np.nan for col in col_ordinais_smae}
        smae_cat_final = np.nanmean(list(smae_cat_col_means.values()))
        metricas_cat['smae_medio'] = {'Q_bar': smae_cat_final, 'ci_lower': np.nan, 'ci_upper': np.nan}

        metricas_cont = {}
        for col in col_continuas:
            metricas_cont[col] = {met: rubin_pool(dict_cont[col][met], n_imp, ALPHA) if dict_cont[col][met] else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan} for met in dict_cont[col]}
        smae_cont_col_means = {col: np.mean(smae_cont_por_coluna.get(col, [])) if smae_cont_por_coluna.get(col) else np.nan for col in col_continuas_smae}
        smae_cont_final = np.nanmean(list(smae_cont_col_means.values()))
        metricas_cont['smae_medio'] = {'Q_bar': smae_cont_final, 'ci_lower': np.nan, 'ci_upper': np.nan}

        # Agregados com Rubin
        nrmse_vals = [dict_cont[col]['nrmse'] for col in col_continuas if dict_cont[col]['nrmse']]
        nmae_vals = [dict_cont[col]['nmae'] for col in col_continuas if dict_cont[col]['nmae']]
        mape_vals = [dict_cont[col]['mape'] for col in col_continuas if dict_cont[col]['mape']]
        smape_vals = [dict_cont[col]['smape'] for col in col_continuas if dict_cont[col]['smape']]
        r2_vals = [dict_cont[col]['r2'] for col in col_continuas if dict_cont[col]['r2']]
        metricas_cont['nrmse_medio'] = rubin_pool([v for sub in nrmse_vals for v in sub], n_imp, ALPHA) if nrmse_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['nmae_medio'] = rubin_pool([v for sub in nmae_vals for v in sub], n_imp, ALPHA) if nmae_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['mape_medio'] = rubin_pool([v for sub in mape_vals for v in sub], n_imp, ALPHA) if mape_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['smape_medio'] = rubin_pool([v for sub in smape_vals for v in sub], n_imp, ALPHA) if smape_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['r2_medio'] = rubin_pool([v for sub in r2_vals for v in sub], n_imp, ALPHA) if r2_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}

        return dict(datasets=datasets, metricas_cat=metricas_cat, metricas_cont=metricas_cont)

# ============================================================================
# BLOCO 7: BOOTSTRAP EXTERNO
# ============================================================================

def bootstrap_misto(df_cod, proporcao_missing, col_continuas, col_ordinais, col_nominais,
                    indices_continuas, indices_ordinais,
                    B=BOOTSTRAP_ITERACOES, m=M_IMPUTATIONS, random_seed=RANDOM_SEED,
                    verbose_first=True, use_conditional_mean=False):
    resultados_bootstrap_cat = []
    resultados_bootstrap_cont = []
    tempos_iteracao = []

    for b in range(1, B+1):
        display(HTML(f'<div class="metrics-highlight">🔷 BOOTSTRAP {b}/{B} - Proporção missing: {proporcao_missing*100:.0f}%</div>'))
        start_iter = time.time()
        rng = np.random.default_rng(random_seed + b)
        idx_boot = rng.choice(df_cod.index, size=len(df_cod), replace=True)
        df_boot_clean = df_cod.loc[idx_boot].reset_index(drop=True)
        df_na, pos_orig = criar_missing_mcar_estratificado(df_boot_clean, proporcao_missing, random_seed=random_seed + b*100)
        show_prog = (verbose_first and b == 1)
        out = multiple_imputation_rubin_misto(
            df_na, pos_orig, df_boot_clean,
            col_continuas, col_ordinais, col_nominais,
            indices_continuas, indices_ordinais,
            m=m, random_seed=random_seed + b*1000, show_progress=show_prog,
            use_conditional_mean=use_conditional_mean)
        cat = out['metricas_cat']
        cont = out['metricas_cont']

        acc = cat.get('acuracia', {}).get('Q_bar', np.nan)
        prec = cat.get('precisao', {}).get('Q_bar', np.nan)
        rec = cat.get('recall', {}).get('Q_bar', np.nan)
        f1 = cat.get('f1_score', {}).get('Q_bar', np.nan)
        kappa = cat.get('kappa', {}).get('Q_bar', np.nan)
        mcc = cat.get('mcc', {}).get('Q_bar', np.nan)
        err_cor = cat.get('erro_correlacao', {}).get('Q_bar', np.nan)
        smae_cat = cat.get('smae_medio', {}).get('Q_bar', np.nan)
        smae_cont = cont.get('smae_medio', {}).get('Q_bar', np.nan)
        gower = cat.get('gower_distance', {}).get('Q_bar', np.nan)
        nrmse = cont.get('nrmse_medio', {}).get('Q_bar', np.nan)
        nmae = cont.get('nmae_medio', {}).get('Q_bar', np.nan)
        mape = cont.get('mape_medio', {}).get('Q_bar', np.nan)
        smape = cont.get('smape_medio', {}).get('Q_bar', np.nan)
        r2 = cont.get('r2_medio', {}).get('Q_bar', np.nan)

        metricas_nomes = ['rmse','mae','mse','cor','nrmse','nmae','mape','smape','r2']
        cont_vals = {}
        for met in metricas_nomes:
            if met in ['nrmse', 'nmae', 'mape', 'smape', 'r2']:
                val = cont.get(f'{met}_medio', {}).get('Q_bar', np.nan)
                vals = [val] if not np.isnan(val) else []
            else:
                vals = []
                for col in col_continuas:
                    if col in cont and met in cont[col]:
                        val = cont[col][met]
                        if isinstance(val, dict):
                            qbar = val.get('Q_bar', np.nan)
                        else:
                            qbar = val if isinstance(val, (int, float)) else np.nan
                        if not np.isnan(qbar):
                            vals.append(qbar)
            cont_vals[met] = np.mean(vals) if vals else np.nan

        elapsed = time.time() - start_iter
        tempos_iteracao.append(elapsed)

        display(HTML(f'''
        <div style="display: grid; grid-template-columns: repeat(5, 1fr); gap: 10px; margin: 10px 0;">
            <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">🎯 Acurácia</div>
                <div style="font-size: 20px; font-weight: bold;">{acc:.2%}</div>
            </div>
            <div style="background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">⚖️ F1</div>
                <div style="font-size: 20px; font-weight: bold;">{f1:.2%}</div>
            </div>
            <div style="background: linear-gradient(135deg, #4facfe 0%, #00f2fe 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">📈 Gower</div>
                <div style="font-size: 20px; font-weight: bold;">{gower:.3f}</div>
            </div>
            <div style="background: linear-gradient(135deg, #11998e 0%, #38ef7d 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">📏 NRMSE</div>
                <div style="font-size: 20px; font-weight: bold;">{nrmse:.3f}</div>
            </div>
            <div style="background: linear-gradient(135deg, #ff6b6b 0%, #ee5a24 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">📐 R²</div>
                <div style="font-size: 20px; font-weight: bold;">{r2:.3f}</div>
            </div>
        </div>
        <div class="info-box">
            ⏱️ Tempo: {elapsed:.2f} segundos |
            RMSE: {cont_vals['rmse']:.3f} | MAE: {cont_vals['mae']:.3f} | NMAE: {cont_vals['nmae']:.4f} |
            MAPE: {cont_vals['mape']:.2f}% | sMAPE: {cont_vals['smape']:.2f}% | Corr: {cont_vals['cor']:.3f} |
            SMAE_Cont: {smae_cont:.3f} | SMAE_Cat: {smae_cat:.3f} | Erro Corr: {err_cor:.4f}
        </div>
        '''))

        resultados_bootstrap_cat.append(dict(
            bootstrap_id=b, acuracia=acc, precisao=prec, recall=rec,
            f1_score=f1, kappa=kappa, mcc=mcc, erro_correlacao=err_cor,
            smae_cat=smae_cat, smae_cont=smae_cont, gower_distance=gower))
        temp_cont = {'bootstrap_id': b}
        temp_cont.update({f'{met}_medio': cont_vals[met] for met in metricas_nomes})
        resultados_bootstrap_cont.append(temp_cont)

    def bootstrap_stats(arr):
        arr = np.array(arr, dtype=float)
        arr = arr[~np.isnan(arr)]
        if len(arr) == 0:
            return dict(media=np.nan, ic_inf=np.nan, ic_sup=np.nan)
        return dict(media=np.mean(arr), ic_inf=np.percentile(arr, 2.5), ic_sup=np.percentile(arr, 97.5))

    df_cat = pd.DataFrame(resultados_bootstrap_cat)
    estat_cat = {met: bootstrap_stats(df_cat[met].values) for met in ['acuracia','precisao','recall','f1_score','kappa','mcc','erro_correlacao','smae_cat','smae_cont','gower_distance']}
    df_cont = pd.DataFrame(resultados_bootstrap_cont)
    estat_cont = {met: bootstrap_stats(df_cont[met].values) for met in df_cont.columns if met != 'bootstrap_id'}
    return dict(estat_cat=estat_cat, estat_cont=estat_cont, tempo_medio=np.mean(tempos_iteracao))

# ============================================================================
# BLOCO 8: EXECUÇÃO PRINCIPAL
# ============================================================================

if __name__ == "__main__":
    display(HTML('<div class="header">🚀 INÍCIO DA EXECUÇÃO - GAUSSIAN COPULA + RUBIN (TIPS 2 CONTÍNUAS + 5 ORDINAIS)</div>'))
    display(HTML(f'<div class="info-box">📅 Data/Hora: {time.strftime("%Y-%m-%d %H:%M:%S")}</div>'))

    USE_CONDITIONAL_MEAN = True
    proporcoes = [0.15, 0.30, 0.40, 0.60, 0.75, 0.90]
    tabela_cat_rows, tabela_cont_rows, tempos_exec = [], [], []

    for idx_prop, prop in enumerate(proporcoes, start=1):
        display(HTML(f'<div class="header">🎯 PROPORÇÃO {idx_prop}/{len(proporcoes)}: {prop*100:.0f}% de missing</div>'))
        inicio = time.time()
        boot_out = bootstrap_misto(
            df_cod, prop, col_continuas, col_ordinais, col_nominais,
            indices_continuas, indices_ordinais,
            B=BOOTSTRAP_ITERACOES, m=M_IMPUTATIONS, random_seed=RANDOM_SEED,
            verbose_first=(prop == proporcoes[0]),
            use_conditional_mean=USE_CONDITIONAL_MEAN
        )
        fim = time.time()
        tempo_total = fim - inicio
        display(HTML(f'<div class="success-box">✅ Proporção {prop*100:.0f}% concluída em {tempo_total:.2f} segundos.</div>'))

        cat = boot_out['estat_cat']
        tabela_cat_rows.append({
            'Missing': f"{prop*100:.0f}%",
            'Acurácia': f"{cat['acuracia']['media']*100:.2f}%",
            'IC_Acurácia': f"[{cat['acuracia']['ic_inf']*100:.2f}%, {cat['acuracia']['ic_sup']*100:.2f}%]",
            'Precisão': f"{cat['precisao']['media']*100:.2f}%",
            'IC_Precisão': f"[{cat['precisao']['ic_inf']*100:.2f}%, {cat['precisao']['ic_sup']*100:.2f}%]",
            'Recall': f"{cat['recall']['media']*100:.2f}%",
            'IC_Recall': f"[{cat['recall']['ic_inf']*100:.2f}%, {cat['recall']['ic_sup']*100:.2f}%]",
            'F1': f"{cat['f1_score']['media']*100:.2f}%",
            'IC_F1': f"[{cat['f1_score']['ic_inf']*100:.2f}%, {cat['f1_score']['ic_sup']*100:.2f}%]",
            'Kappa': f"{cat['kappa']['media']:.3f}",
            'IC_Kappa': f"[{cat['kappa']['ic_inf']:.3f}, {cat['kappa']['ic_sup']:.3f}]",
            'MCC': f"{cat['mcc']['media']:.3f}",
            'IC_MCC': f"[{cat['mcc']['ic_inf']:.3f}, {cat['mcc']['ic_sup']:.3f}]",
            'Erro_Correlação': f"{cat['erro_correlacao']['media']:.4f}",
            'IC_ErroCor': f"[{cat['erro_correlacao']['ic_inf']:.4f}, {cat['erro_correlacao']['ic_sup']:.4f}]",
            'SMAE_Categórico': f"{cat['smae_cat']['media']:.3f}",
            'IC_SMAE_Cat': f"[{cat['smae_cat']['ic_inf']:.3f}, {cat['smae_cat']['ic_sup']:.3f}]",
            'SMAE_Contínuo': f"{cat['smae_cont']['media']:.3f}",
            'IC_SMAE_Cont': f"[{cat['smae_cont']['ic_inf']:.3f}, {cat['smae_cont']['ic_sup']:.3f}]",
            'Gower_Distance': f"{cat['gower_distance']['media']:.4f}",
            'IC_Gower': f"[{cat['gower_distance']['ic_inf']:.4f}, {cat['gower_distance']['ic_sup']:.4f}]"
        })

        cont = boot_out['estat_cont']
        tabela_cont_rows.append({
            'Missing': f"{prop*100:.0f}%",
            'RMSE': f"{cont['rmse_medio']['media']:.3f}",
            'IC_RMSE': f"[{cont['rmse_medio']['ic_inf']:.3f}, {cont['rmse_medio']['ic_sup']:.3f}]",
            'MAE': f"{cont['mae_medio']['media']:.3f}",
            'IC_MAE': f"[{cont['mae_medio']['ic_inf']:.3f}, {cont['mae_medio']['ic_sup']:.3f}]",
            'NMAE': f"{cont['nmae_medio']['media']:.4f}",
            'IC_NMAE': f"[{cont['nmae_medio']['ic_inf']:.4f}, {cont['nmae_medio']['ic_sup']:.4f}]",
            'MSE': f"{cont['mse_medio']['media']:.3f}",
            'IC_MSE': f"[{cont['mse_medio']['ic_inf']:.3f}, {cont['mse_medio']['ic_sup']:.3f}]",
            'MAPE(%)': f"{cont['mape_medio']['media']:.2f}",
            'IC_MAPE': f"[{cont['mape_medio']['ic_inf']:.2f}, {cont['mape_medio']['ic_sup']:.2f}]",
            'sMAPE(%)': f"{cont['smape_medio']['media']:.2f}",
            'IC_sMAPE': f"[{cont['smape_medio']['ic_inf']:.2f}, {cont['smape_medio']['ic_sup']:.2f}]",
            'Correlação': f"{cont['cor_medio']['media']:.3f}",
            'IC_Cor': f"[{cont['cor_medio']['ic_inf']:.3f}, {cont['cor_medio']['ic_sup']:.3f}]",
            'NRMSE': f"{cont['nrmse_medio']['media']:.4f}",
            'IC_NRMSE': f"[{cont['nrmse_medio']['ic_inf']:.4f}, {cont['nrmse_medio']['ic_sup']:.4f}]",
            'R²': f"{cont['r2_medio']['media']:.4f}",
            'IC_R²': f"[{cont['r2_medio']['ic_inf']:.4f}, {cont['r2_medio']['ic_sup']:.4f}]"
        })

        tempos_exec.append({
            'Proporção': f"{prop*100:.0f}%",
            'Tempo_Total': f"{tempo_total:.2f}",
            'Tempo_Medio_Boot': f"{boot_out['tempo_medio']:.2f}"
        })

    display(HTML('<div class="header">📊 RESULTADOS FINAIS - VARIÁVEIS ORDINAIS (size, day, time, smoker, sex)</div>'))
    df_cat_final = pd.DataFrame(tabela_cat_rows)
    display(df_cat_final.style.set_table_attributes('class="dataframe-table"'))

    display(HTML('<div class="header">📊 RESULTADOS FINAIS - VARIÁVEIS CONTÍNUAS (total_bill, tip)</div>'))
    df_cont_final = pd.DataFrame(tabela_cont_rows)
    display(df_cont_final.style.set_table_attributes('class="dataframe-table"'))

    display(HTML('<div class="header">⏱️ TEMPOS DE EXECUÇÃO POR PROPORÇÃO</div>'))
    df_tempo = pd.DataFrame(tempos_exec)
    display(df_tempo.style.set_table_attributes('class="dataframe-table"'))

    try:
        df_plot = pd.DataFrame(tabela_cat_rows)
        df_plot['prop'] = df_plot['Missing'].str.replace('%','').astype(float)/100
        df_plot['Acc'] = df_plot['Acurácia'].str.replace('%','').astype(float)/100
        plt.figure(figsize=(8,5))
        sns.lineplot(x='prop', y='Acc', data=df_plot, marker='o', color='blue')
        plt.xlabel("Proporção de missing")
        plt.ylabel("Acurácia")
        plt.title("Acurácia por proporção de dados faltantes - Dataset Tips (2 cont, 5 ord)")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()
    except Exception as e:
        display(HTML(f'<div class="warning-box">⚠️ Gráfico não pôde ser gerado: {e}</div>'))

    display(HTML('<div class="success-box">🎉 EXECUÇÃO CONCLUÍDA - TIPS COM 2 CONTÍNUAS E 5 ORDINAIS!</div>'))
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
================================================================================
GAUSSIAN COPULA + BOOTSTRAP PARA DATASET TIPS (Restaurant Tips)
Versão com 2 contínuas e 5 ordinais (size, day, time, smoker, sex)
================================================================================
"""

# ============================================================================
# BLOCO 1: IMPORTAÇÕES E CONFIGURAÇÕES INICIAIS
# ============================================================================

import subprocess
import sys
import warnings
import time
import random
from IPython.display import display, HTML

CSS_STYLES = """
<style>
    .header {
        background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        color: white;
        padding: 15px;
        border-radius: 10px;
        margin: 10px 0;
        text-align: center;
        font-family: Arial, sans-serif;
        font-weight: bold;
        font-size: 18px;
    }
    .success-box {
        background-color: #d4edda;
        color: #155724;
        padding: 12px;
        border: 1px solid #c3e6cb;
        border-radius: 5px;
        margin: 10px 0;
        font-weight: bold;
    }
    .warning-box {
        background-color: #fff3cd;
        color: #856404;
        padding: 12px;
        border: 1px solid #ffeaa7;
        border-radius: 5px;
        margin: 10px 0;
        font-weight: bold;
    }
    .info-box {
        background-color: #d1ecf1;
        color: #0c5460;
        padding: 12px;
        border: 1px solid #bee5eb;
        border-radius: 5px;
        margin: 10px 0;
    }
    .metrics-highlight {
        background: linear-gradient(135deg, #ffd89b 0%, #19547b 100%);
        color: white;
        padding: 15px;
        border-radius: 10px;
        margin: 10px 0;
        text-align: center;
        font-weight: bold;
    }
    .dataframe-table {
        font-family: Arial, sans-serif;
        border-collapse: collapse;
        width: 100%;
        box-shadow: 0 4px 8px rgba(0,0,0,0.1);
        border-radius: 10px;
        overflow: hidden;
        margin: 20px 0;
    }
    .dataframe-table th {
        background: linear-gradient(135deg, #2c3e50 0%, #3498db 100%);
        color: white;
        padding: 12px;
        text-align: left;
        font-weight: bold;
    }
    .dataframe-table td {
        padding: 10px;
        border-bottom: 1px solid #ddd;
    }
    .dataframe-table tr:nth-child(even) {
        background-color: #f8f9fa;
    }
    .dataframe-table tr:hover {
        background-color: #e9ecef;
        transform: scale(1.01);
        transition: all 0.2s ease;
    }
    .rubin-rules-box {
        background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        color: white;
        padding: 15px;
        border-radius: 10px;
        margin: 10px 0;
        text-align: center;
        font-weight: bold;
    }
    .iteration-box {
        background-color: #e8f4fd;
        border-left: 5px solid #3498db;
        padding: 10px;
        margin: 10px 0;
        border-radius: 5px;
    }
    .encoding-badge {
        background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%);
        color: white;
        padding: 5px 10px;
        border-radius: 15px;
        font-size: 12px;
        font-weight: bold;
        display: inline-block;
        margin: 2px;
    }
</style>
"""
display(HTML(CSS_STYLES))

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

for lib in ['numpy', 'pandas', 'scipy', 'scikit-learn', 'matplotlib', 'seaborn']:
    try:
        __import__(lib)
    except ImportError:
        install(lib)

import numpy as np
import pandas as pd
from scipy import stats
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, cohen_kappa_score, matthews_corrcoef)
import matplotlib.pyplot as plt
import seaborn as sns

try:
    from gcimpute.gaussian_copula import GaussianCopula
except ImportError:
    print("Instalando gcimpute do GitHub...")
    subprocess.check_call([sys.executable, "-m", "pip", "install",
                           "git+https://github.com/udellgroup/gcimpute.git"])
    from gcimpute.gaussian_copula import GaussianCopula

warnings.filterwarnings("ignore")
if not hasattr(np, 'round_'):
    np.round_ = np.round

# ============================================================================
# BLOCO 2: CONFIGURAÇÕES GLOBAIS
# ============================================================================

RANDOM_SEED = 42
BOOTSTRAP_ITERACOES = 100
M_IMPUTATIONS = 20
ALPHA = 0.05
MAX_ITER_EM = 200
TOL_EM = 0.001
np.random.seed(RANDOM_SEED)

# ============================================================================
# BLOCO 3: CARREGAMENTO E PRÉ-PROCESSAMENTO (TIPS) - 2 CONTÍNUAS, 5 ORDINAIS
# ============================================================================

display(HTML('<div class="header">🍽️ CARREGANDO DATASET TIPS (2 contínuas, 5 ordinais)</div>'))

tips = sns.load_dataset("tips")
df_raw = tips.copy()
display(HTML(f'<div class="success-box">✅ Dados carregados: {df_raw.shape[0]} linhas, {df_raw.shape[1]} colunas.</div>'))
display(HTML('<div class="info-box"><b>Primeiras linhas:</b></div>'))
display(df_raw.head().style.set_table_attributes('class="dataframe-table"'))

df_clean = df_raw.copy()

# Definição dos tipos
col_continuas = ["total_bill", "tip"]
col_ordinais = ["size", "day", "time", "smoker", "sex"]
col_nominais = []

# Define ordem para as categóricas (exceto size, que já é int)
df_clean['day'] = pd.Categorical(df_clean['day'], categories=['Thur','Fri','Sat','Sun'], ordered=True)
df_clean['time'] = pd.Categorical(df_clean['time'], categories=['Lunch','Dinner'], ordered=True)
df_clean['smoker'] = pd.Categorical(df_clean['smoker'], categories=['No','Yes'], ordered=True)
df_clean['sex'] = pd.Categorical(df_clean['sex'], categories=['Female','Male'], ordered=True)
# size permanece int

niveis_originais = {}
for col in col_ordinais + col_nominais:
    df_clean[col] = df_clean[col].astype("category")
    niveis_originais[col] = list(df_clean[col].cat.categories)

display(HTML('<div class="info-box"><b>Níveis das variáveis ordinais (com ordem):</b></div>'))
for col, lvls in niveis_originais.items():
    display(HTML(f'<span class="encoding-badge">{col}: {lvls}</span> '))

# Codificação para inteiros (preserva ordem)
df_cod = df_clean.copy()
for col in col_ordinais + col_nominais:
    df_cod[col] = df_cod[col].cat.codes

display(HTML('<div class="info-box"><b>Dados codificados (head):</b></div>'))
display(df_cod.head().style.set_table_attributes('class="dataframe-table"'))

colunas_ordenadas = list(df_cod.columns)
indices_continuas = [colunas_ordenadas.index(col) for col in col_continuas]
indices_ordinais = [colunas_ordenadas.index(col) for col in (col_ordinais + col_nominais)]

print("Índices contínuas:", indices_continuas)
print("Índices ordinais :", indices_ordinais)

# ============================================================================
# BLOCO 4: FUNÇÕES AUXILIARES (MISSING, IMPUTAÇÃO, MÉTRICAS)
# ============================================================================

def criar_missing_mcar_estratificado(df, percentual=0.75, random_seed=42):
    np.random.seed(random_seed)
    random.seed(random_seed)
    df_com_na = df.copy()
    posicoes_originais = {}
    for col in df.columns:
        n_missing = int(len(df) * percentual)
        indices = np.random.choice(df.index, size=n_missing, replace=False)
        for i in indices:
            posicoes_originais[(i, col)] = df.loc[i, col]
            df_com_na.loc[i, col] = np.nan
    return df_com_na, posicoes_originais

def imputacoes_gcimpute(df_na, indices_continuas, indices_ordinais,
                        m=M_IMPUTATIONS, random_seed=None, verbose=True,
                        maxit=MAX_ITER_EM, tol=TOL_EM,
                        use_conditional_mean=False):
    X = df_na.astype(float).values
    model = GaussianCopula(
        max_iter=maxit, tol=tol,
        random_state=random_seed if random_seed is not None else RANDOM_SEED,
        verbose=1 if verbose else 0,
    )
    if verbose:
        print(f"  Ajustando Gaussian Copula (EM)...")
        print(f"    contínuas: {indices_continuas}")
        print(f"    ordinais : {indices_ordinais}")
    model.fit(X, continuous=indices_continuas, ordinal=indices_ordinais)
    if use_conditional_mean:
        X_imp = model.transform(X)
        df_imp = pd.DataFrame(X_imp, columns=df_na.columns, index=df_na.index)
        datasets = [df_imp]
    else:
        X_imp = model.sample_imputation(X=X, num=m)
        datasets = [pd.DataFrame(X_imp[:, :, i], columns=df_na.columns, index=df_na.index) for i in range(m)]
    return datasets

def metricas_continuas(reais, imputados):
    mask = ~np.isnan(reais) & ~np.isnan(imputados)
    if mask.sum() < 2:
        return dict(rmse=np.nan, mae=np.nan, mse=np.nan, cor=np.nan, nrmse=np.nan,
                    nmae=np.nan, mape=np.nan, smape=np.nan, r2=np.nan)
    r = reais[mask]
    p = imputados[mask]
    mse = np.mean((r - p) ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(r - p))
    cor = np.corrcoef(r, p)[0, 1] if len(r) > 1 else np.nan
    intervalo = np.ptp(r)
    nrmse = rmse / intervalo if intervalo > 0 else np.nan
    nmae = mae / intervalo if intervalo > 0 else np.nan
    with np.errstate(divide='ignore', invalid='ignore'):
        mape = np.mean(np.abs((r - p) / (r + 1e-8))) * 100
        mape = mape if np.isfinite(mape) else np.nan
    with np.errstate(divide='ignore', invalid='ignore'):
        denom = (np.abs(r) + np.abs(p) + 1e-8)
        smape = np.mean(2 * np.abs(r - p) / denom) * 100
        smape = smape if np.isfinite(smape) else np.nan
    ss_res = np.sum((r - p) ** 2)
    ss_tot = np.sum((r - np.mean(r)) ** 2)
    r2 = 1 - (ss_res / ss_tot) if ss_tot > 0 else np.nan
    return dict(rmse=rmse, mae=mae, mse=mse, cor=cor, nrmse=nrmse,
                nmae=nmae, mape=mape, smape=smape, r2=r2)

def erro_matriz_correlacao(df_original, df_imputado, col_continuas):
    orig = df_original[col_continuas].dropna()
    imp = df_imputado.loc[orig.index, col_continuas]
    cor_orig = orig.corr().values
    cor_imp = imp.corr().values
    diff = np.linalg.norm(cor_imp - cor_orig, 'fro')
    norm_orig = np.linalg.norm(cor_orig, 'fro')
    return diff / norm_orig if norm_orig > 0 else np.nan

def calcular_smae_por_coluna(valores_reais, valores_imputados, referencia):
    if len(valores_reais) == 0:
        return np.nan
    reais = np.array(valores_reais)
    imputados = np.array(valores_imputados)
    erro_imp = np.sum(np.abs(imputados - reais))
    erro_ref = np.sum(np.abs(referencia - reais))
    if erro_ref == 0:
        return np.nan
    return erro_imp / erro_ref

def gower_distance_matrix(df_real, df_imputado, col_continuas, col_categoricas):
    n = len(df_real)
    if n == 0:
        return np.nan
    all_cols = col_continuas + col_categoricas
    df_real_sub = df_real[all_cols].reset_index(drop=True)
    df_imp_sub = df_imputado[all_cols].reset_index(drop=True)
    gower_distances = np.zeros(n)
    for col in col_continuas:
        real_vals = df_real_sub[col].values
        imp_vals = df_imp_sub[col].values
        rng = np.ptp(real_vals)
        if rng == 0:
            rng = 1.0
        diff = np.abs(real_vals - imp_vals) / rng
        gower_distances += diff
    for col in col_categoricas:
        real_vals = df_real_sub[col].values
        imp_vals = df_imp_sub[col].values
        diff = (real_vals != imp_vals).astype(float)
        gower_distances += diff
    n_cols = len(all_cols)
    gower_distances /= n_cols
    return np.mean(gower_distances)

# ============================================================================
# BLOCO 5: REGRAS DE RUBIN
# ============================================================================

def rubin_pool(estimativas, m, alpha=0.05):
    estimativas = np.array(estimativas, dtype=float)
    estimativas = estimativas[~np.isnan(estimativas)]
    k = len(estimativas)
    if k == 0:
        return {'Q_bar': np.nan, 'se': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
    Q_bar = np.mean(estimativas)
    if k == 1:
        return {'Q_bar': Q_bar, 'se': 0.0, 'ci_lower': Q_bar, 'ci_upper': Q_bar}
    B = np.var(estimativas, ddof=1)
    T = (1 + 1/m) * B
    se = np.sqrt(T)
    gl = m - 1
    t_val = stats.t.ppf(1 - alpha/2, gl)
    ci_lower = Q_bar - t_val * se
    ci_upper = Q_bar + t_val * se
    return {'Q_bar': Q_bar, 'se': se, 'ci_lower': ci_lower, 'ci_upper': ci_upper}

# ============================================================================
# BLOCO 6: FUNÇÃO PRINCIPAL DE IMPUTAÇÃO + AVALIAÇÃO
# ============================================================================

def multiple_imputation_rubin_misto(df_na, posicoes_originais, df_original_clean,
                                    col_continuas, col_ordinais, col_nominais,
                                    indices_continuas, indices_ordinais,
                                    m=M_IMPUTATIONS, random_seed=RANDOM_SEED, show_progress=True,
                                    use_conditional_mean=False):
    # SMAE sobre contínuas e ordinais (todas as ordinais)
    col_continuas_smae = col_continuas                     # total_bill, tip
    col_ordinais_smae = col_ordinais                       # size, day, time, smoker, sex

    if show_progress:
        if use_conditional_mean:
            display(HTML(f'<div class="iteration-box">🔹 Gerando imputação por média condicional...</div>'))
        else:
            display(HTML(f'<div class="iteration-box">🔹 Gerando {m} imputações via gcimpute...</div>'))
    datasets = imputacoes_gcimpute(df_na, indices_continuas, indices_ordinais,
                                   m=m, random_seed=random_seed, verbose=show_progress,
                                   use_conditional_mean=use_conditional_mean)
    n_imp = len(datasets)
    if n_imp == 0:
        return dict(datasets=[], metricas_cat={}, metricas_cont={})

    medianas_cont = {col: df_original_clean[col].median() for col in col_continuas_smae}
    modas_cat = {}
    for col in col_ordinais_smae:
        moda = df_original_clean[col].mode()
        modas_cat[col] = moda.iloc[0] if len(moda) > 0 else np.nan

    acc_list, prec_list, rec_list, f1_list, kappa_list, mcc_list = [], [], [], [], [], []
    erro_cor_list = []
    dict_cont = {col: {met: [] for met in ['rmse','mae','mse','cor','nrmse','nmae','mape','smape','r2']} for col in col_continuas}
    smae_cont_por_coluna = {col: [] for col in col_continuas_smae}
    smae_cat_por_coluna = {col: [] for col in col_ordinais_smae}
    cols_categoricas = col_ordinais + col_nominais   # todas as categóricas (aqui são apenas ordinais)
    gower_list = []

    for i, df_imp in enumerate(datasets):
        if show_progress:
            display(HTML(f'<div class="metrics-highlight">📊 Avaliando imputação {i+1}/{n_imp}</div>'))
        err_cor = erro_matriz_correlacao(df_original_clean, df_imp, col_continuas)
        erro_cor_list.append(err_cor)

        # Avaliação categórica (todas as ordinais)
        reais_cat, preditos_cat = [], []
        reais_cat_por_col = {col: [] for col in cols_categoricas}
        imps_cat_por_col = {col: [] for col in cols_categoricas}
        for (linha, coluna), valor_real in posicoes_originais.items():
            if coluna in cols_categoricas:
                valor_imp = df_imp.at[linha, coluna]
                if pd.notna(valor_real) and pd.notna(valor_imp):
                    pred = int(round(valor_imp))
                    real = int(valor_real)
                    k = len(niveis_originais.get(coluna, [0]))
                    pred = max(0, min(pred, k-1))
                    reais_cat.append(real)
                    preditos_cat.append(pred)
                    reais_cat_por_col[coluna].append(real)
                    imps_cat_por_col[coluna].append(pred)
        if reais_cat:
            try:
                acc_list.append(accuracy_score(reais_cat, preditos_cat))
                prec_list.append(precision_score(reais_cat, preditos_cat, average='macro', zero_division=0))
                rec_list.append(recall_score(reais_cat, preditos_cat, average='macro', zero_division=0))
                f1_list.append(f1_score(reais_cat, preditos_cat, average='macro', zero_division=0))
                kappa_list.append(cohen_kappa_score(reais_cat, preditos_cat))
                mcc_list.append(matthews_corrcoef(reais_cat, preditos_cat))
            except:
                pass
            # SMAE ordinal (para cada coluna ordinal)
            for col in col_ordinais_smae:
                if col in reais_cat_por_col and reais_cat_por_col[col]:
                    smae = calcular_smae_por_coluna(reais_cat_por_col[col], imps_cat_por_col[col], modas_cat[col])
                    smae_cat_por_coluna[col].append(smae)

        # Avaliação contínua
        for col in col_continuas:
            reais_cont, imps_cont = [], []
            for (linha, coluna), valor_real in posicoes_originais.items():
                if coluna == col:
                    imp_val = df_imp.at[linha, col]
                    if pd.notna(valor_real) and pd.notna(imp_val):
                        reais_cont.append(valor_real)
                        imps_cont.append(imp_val)
            if reais_cont:
                met = metricas_continuas(np.array(reais_cont), np.array(imps_cont))
                for mname in dict_cont[col]:
                    dict_cont[col][mname].append(met[mname])
                if col in col_continuas_smae:
                    smae = calcular_smae_por_coluna(reais_cont, imps_cont, medianas_cont[col])
                    smae_cont_por_coluna[col].append(smae)

        # Distância de Gower
        real_dict = {}
        imp_dict = {}
        for (linha, coluna), val_real in posicoes_originais.items():
            if linha not in real_dict:
                real_dict[linha] = {}
                imp_dict[linha] = {}
            real_dict[linha][coluna] = val_real
            imp_dict[linha][coluna] = df_imp.at[linha, coluna]
        real_rows, imp_rows = [], []
        for linha in real_dict.keys():
            row_real = {}
            row_imp = {}
            for col in col_continuas + cols_categoricas:
                if col in real_dict[linha]:
                    row_real[col] = real_dict[linha][col]
                    row_imp[col] = imp_dict[linha][col]
                else:
                    row_real[col] = df_original_clean.at[linha, col]
                    row_imp[col] = df_original_clean.at[linha, col]
            real_rows.append(row_real)
            imp_rows.append(row_imp)
        df_real_gower = pd.DataFrame(real_rows)
        df_imp_gower = pd.DataFrame(imp_rows)
        gower_list.append(gower_distance_matrix(df_real_gower, df_imp_gower, col_continuas, cols_categoricas))

    # Agregação
    if use_conditional_mean:
        metricas_cat = {
            'acuracia': acc_list[0] if acc_list else np.nan,
            'precisao': prec_list[0] if prec_list else np.nan,
            'recall': rec_list[0] if rec_list else np.nan,
            'f1_score': f1_list[0] if f1_list else np.nan,
            'kappa': kappa_list[0] if kappa_list else np.nan,
            'mcc': mcc_list[0] if mcc_list else np.nan,
            'erro_correlacao': erro_cor_list[0] if erro_cor_list else np.nan,
            'gower_distance': gower_list[0] if gower_list else np.nan
        }
        smae_cat_col_means = {col: np.mean(smae_cat_por_coluna.get(col, [])) if smae_cat_por_coluna.get(col) else np.nan for col in col_ordinais_smae}
        metricas_cat['smae_medio'] = np.nanmean(list(smae_cat_col_means.values()))

        metricas_cont = {}
        for col in col_continuas:
            metricas_cont[col] = {met: dict_cont[col][met][0] if dict_cont[col][met] else np.nan for met in dict_cont[col]}
        smae_cont_col_means = {col: np.mean(smae_cont_por_coluna.get(col, [])) if smae_cont_por_coluna.get(col) else np.nan for col in col_continuas_smae}
        metricas_cont['smae_medio'] = np.nanmean(list(smae_cont_col_means.values()))
        # Agregados
        metricas_cont['nrmse_medio'] = np.mean([dict_cont[col]['nrmse'][0] for col in col_continuas if dict_cont[col]['nrmse']]) if any(dict_cont[col]['nrmse'] for col in col_continuas) else np.nan
        metricas_cont['nmae_medio'] = np.mean([dict_cont[col]['nmae'][0] for col in col_continuas if dict_cont[col]['nmae']]) if any(dict_cont[col]['nmae'] for col in col_continuas) else np.nan
        metricas_cont['mape_medio'] = np.mean([dict_cont[col]['mape'][0] for col in col_continuas if dict_cont[col]['mape']]) if any(dict_cont[col]['mape'] for col in col_continuas) else np.nan
        metricas_cont['smape_medio'] = np.mean([dict_cont[col]['smape'][0] for col in col_continuas if dict_cont[col]['smape']]) if any(dict_cont[col]['smape'] for col in col_continuas) else np.nan
        metricas_cont['r2_medio'] = np.mean([dict_cont[col]['r2'][0] for col in col_continuas if dict_cont[col]['r2']]) if any(dict_cont[col]['r2'] for col in col_continuas) else np.nan

        metricas_cat_rubin = {k: {'Q_bar': v, 'ci_lower': np.nan, 'ci_upper': np.nan} for k, v in metricas_cat.items()}
        metricas_cont_rubin = {}
        for k, v in metricas_cont.items():
            if isinstance(v, dict):
                metricas_cont_rubin[k] = {kk: vv for kk, vv in v.items()}
            else:
                metricas_cont_rubin[k] = {'Q_bar': v, 'ci_lower': np.nan, 'ci_upper': np.nan}
        return dict(datasets=datasets, metricas_cat=metricas_cat_rubin, metricas_cont=metricas_cont_rubin)
    else:
        # Múltiplas imputações com Rubin
        metricas_cat = {}
        for nome, lst in [('acuracia', acc_list), ('precisao', prec_list), ('recall', rec_list),
                          ('f1_score', f1_list), ('kappa', kappa_list), ('mcc', mcc_list)]:
            metricas_cat[nome] = rubin_pool(lst, n_imp, ALPHA) if lst else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cat['erro_correlacao'] = rubin_pool(erro_cor_list, n_imp, ALPHA) if erro_cor_list else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cat['gower_distance'] = rubin_pool(gower_list, n_imp, ALPHA) if gower_list else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        smae_cat_col_means = {col: np.mean(smae_cat_por_coluna.get(col, [])) if smae_cat_por_coluna.get(col) else np.nan for col in col_ordinais_smae}
        smae_cat_final = np.nanmean(list(smae_cat_col_means.values()))
        metricas_cat['smae_medio'] = {'Q_bar': smae_cat_final, 'ci_lower': np.nan, 'ci_upper': np.nan}

        metricas_cont = {}
        for col in col_continuas:
            metricas_cont[col] = {met: rubin_pool(dict_cont[col][met], n_imp, ALPHA) if dict_cont[col][met] else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan} for met in dict_cont[col]}
        smae_cont_col_means = {col: np.mean(smae_cont_por_coluna.get(col, [])) if smae_cont_por_coluna.get(col) else np.nan for col in col_continuas_smae}
        smae_cont_final = np.nanmean(list(smae_cont_col_means.values()))
        metricas_cont['smae_medio'] = {'Q_bar': smae_cont_final, 'ci_lower': np.nan, 'ci_upper': np.nan}

        # Agregados com Rubin
        nrmse_vals = [dict_cont[col]['nrmse'] for col in col_continuas if dict_cont[col]['nrmse']]
        nmae_vals = [dict_cont[col]['nmae'] for col in col_continuas if dict_cont[col]['nmae']]
        mape_vals = [dict_cont[col]['mape'] for col in col_continuas if dict_cont[col]['mape']]
        smape_vals = [dict_cont[col]['smape'] for col in col_continuas if dict_cont[col]['smape']]
        r2_vals = [dict_cont[col]['r2'] for col in col_continuas if dict_cont[col]['r2']]
        metricas_cont['nrmse_medio'] = rubin_pool([v for sub in nrmse_vals for v in sub], n_imp, ALPHA) if nrmse_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['nmae_medio'] = rubin_pool([v for sub in nmae_vals for v in sub], n_imp, ALPHA) if nmae_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['mape_medio'] = rubin_pool([v for sub in mape_vals for v in sub], n_imp, ALPHA) if mape_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['smape_medio'] = rubin_pool([v for sub in smape_vals for v in sub], n_imp, ALPHA) if smape_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}
        metricas_cont['r2_medio'] = rubin_pool([v for sub in r2_vals for v in sub], n_imp, ALPHA) if r2_vals else {'Q_bar': np.nan, 'ci_lower': np.nan, 'ci_upper': np.nan}

        return dict(datasets=datasets, metricas_cat=metricas_cat, metricas_cont=metricas_cont)

# ============================================================================
# BLOCO 7: BOOTSTRAP EXTERNO
# ============================================================================

def bootstrap_misto(df_cod, proporcao_missing, col_continuas, col_ordinais, col_nominais,
                    indices_continuas, indices_ordinais,
                    B=BOOTSTRAP_ITERACOES, m=M_IMPUTATIONS, random_seed=RANDOM_SEED,
                    verbose_first=True, use_conditional_mean=False):
    resultados_bootstrap_cat = []
    resultados_bootstrap_cont = []
    tempos_iteracao = []

    for b in range(1, B+1):
        display(HTML(f'<div class="metrics-highlight">🔷 BOOTSTRAP {b}/{B} - Proporção missing: {proporcao_missing*100:.0f}%</div>'))
        start_iter = time.time()
        rng = np.random.default_rng(random_seed + b)
        idx_boot = rng.choice(df_cod.index, size=len(df_cod), replace=True)
        df_boot_clean = df_cod.loc[idx_boot].reset_index(drop=True)
        df_na, pos_orig = criar_missing_mcar_estratificado(df_boot_clean, proporcao_missing, random_seed=random_seed + b*100)
        show_prog = (verbose_first and b == 1)
        out = multiple_imputation_rubin_misto(
            df_na, pos_orig, df_boot_clean,
            col_continuas, col_ordinais, col_nominais,
            indices_continuas, indices_ordinais,
            m=m, random_seed=random_seed + b*1000, show_progress=show_prog,
            use_conditional_mean=use_conditional_mean)
        cat = out['metricas_cat']
        cont = out['metricas_cont']

        acc = cat.get('acuracia', {}).get('Q_bar', np.nan)
        prec = cat.get('precisao', {}).get('Q_bar', np.nan)
        rec = cat.get('recall', {}).get('Q_bar', np.nan)
        f1 = cat.get('f1_score', {}).get('Q_bar', np.nan)
        kappa = cat.get('kappa', {}).get('Q_bar', np.nan)
        mcc = cat.get('mcc', {}).get('Q_bar', np.nan)
        err_cor = cat.get('erro_correlacao', {}).get('Q_bar', np.nan)
        smae_cat = cat.get('smae_medio', {}).get('Q_bar', np.nan)
        smae_cont = cont.get('smae_medio', {}).get('Q_bar', np.nan)
        gower = cat.get('gower_distance', {}).get('Q_bar', np.nan)
        nrmse = cont.get('nrmse_medio', {}).get('Q_bar', np.nan)
        nmae = cont.get('nmae_medio', {}).get('Q_bar', np.nan)
        mape = cont.get('mape_medio', {}).get('Q_bar', np.nan)
        smape = cont.get('smape_medio', {}).get('Q_bar', np.nan)
        r2 = cont.get('r2_medio', {}).get('Q_bar', np.nan)

        metricas_nomes = ['rmse','mae','mse','cor','nrmse','nmae','mape','smape','r2']
        cont_vals = {}
        for met in metricas_nomes:
            if met in ['nrmse', 'nmae', 'mape', 'smape', 'r2']:
                val = cont.get(f'{met}_medio', {}).get('Q_bar', np.nan)
                vals = [val] if not np.isnan(val) else []
            else:
                vals = []
                for col in col_continuas:
                    if col in cont and met in cont[col]:
                        val = cont[col][met]
                        if isinstance(val, dict):
                            qbar = val.get('Q_bar', np.nan)
                        else:
                            qbar = val if isinstance(val, (int, float)) else np.nan
                        if not np.isnan(qbar):
                            vals.append(qbar)
            cont_vals[met] = np.mean(vals) if vals else np.nan

        elapsed = time.time() - start_iter
        tempos_iteracao.append(elapsed)

        display(HTML(f'''
        <div style="display: grid; grid-template-columns: repeat(5, 1fr); gap: 10px; margin: 10px 0;">
            <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">🎯 Acurácia</div>
                <div style="font-size: 20px; font-weight: bold;">{acc:.2%}</div>
            </div>
            <div style="background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">⚖️ F1</div>
                <div style="font-size: 20px; font-weight: bold;">{f1:.2%}</div>
            </div>
            <div style="background: linear-gradient(135deg, #4facfe 0%, #00f2fe 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">📈 Gower</div>
                <div style="font-size: 20px; font-weight: bold;">{gower:.3f}</div>
            </div>
            <div style="background: linear-gradient(135deg, #11998e 0%, #38ef7d 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">📏 NRMSE</div>
                <div style="font-size: 20px; font-weight: bold;">{nrmse:.3f}</div>
            </div>
            <div style="background: linear-gradient(135deg, #ff6b6b 0%, #ee5a24 100%); color: white; padding: 10px; border-radius: 8px; text-align: center;">
                <div style="font-size: 12px;">📐 R²</div>
                <div style="font-size: 20px; font-weight: bold;">{r2:.3f}</div>
            </div>
        </div>
        <div class="info-box">
            ⏱️ Tempo: {elapsed:.2f} segundos |
            RMSE: {cont_vals['rmse']:.3f} | MAE: {cont_vals['mae']:.3f} | NMAE: {cont_vals['nmae']:.4f} |
            MAPE: {cont_vals['mape']:.2f}% | sMAPE: {cont_vals['smape']:.2f}% | Corr: {cont_vals['cor']:.3f} |
            SMAE_Cont: {smae_cont:.3f} | SMAE_Cat: {smae_cat:.3f} | Erro Corr: {err_cor:.4f}
        </div>
        '''))

        resultados_bootstrap_cat.append(dict(
            bootstrap_id=b, acuracia=acc, precisao=prec, recall=rec,
            f1_score=f1, kappa=kappa, mcc=mcc, erro_correlacao=err_cor,
            smae_cat=smae_cat, smae_cont=smae_cont, gower_distance=gower))
        temp_cont = {'bootstrap_id': b}
        temp_cont.update({f'{met}_medio': cont_vals[met] for met in metricas_nomes})
        resultados_bootstrap_cont.append(temp_cont)

    def bootstrap_stats(arr):
        arr = np.array(arr, dtype=float)
        arr = arr[~np.isnan(arr)]
        if len(arr) == 0:
            return dict(media=np.nan, ic_inf=np.nan, ic_sup=np.nan)
        return dict(media=np.mean(arr), ic_inf=np.percentile(arr, 2.5), ic_sup=np.percentile(arr, 97.5))

    df_cat = pd.DataFrame(resultados_bootstrap_cat)
    estat_cat = {met: bootstrap_stats(df_cat[met].values) for met in ['acuracia','precisao','recall','f1_score','kappa','mcc','erro_correlacao','smae_cat','smae_cont','gower_distance']}
    df_cont = pd.DataFrame(resultados_bootstrap_cont)
    estat_cont = {met: bootstrap_stats(df_cont[met].values) for met in df_cont.columns if met != 'bootstrap_id'}
    return dict(estat_cat=estat_cat, estat_cont=estat_cont, tempo_medio=np.mean(tempos_iteracao))

# ============================================================================
# BLOCO 8: EXECUÇÃO PRINCIPAL
# ============================================================================

if __name__ == "__main__":
    display(HTML('<div class="header">🚀 INÍCIO DA EXECUÇÃO - GAUSSIAN COPULA + RUBIN (TIPS 2 CONTÍNUAS + 5 ORDINAIS)</div>'))
    display(HTML(f'<div class="info-box">📅 Data/Hora: {time.strftime("%Y-%m-%d %H:%M:%S")}</div>'))

    USE_CONDITIONAL_MEAN = True
    proporcoes = [0.15, 0.30, 0.40, 0.60, 0.75, 0.90]
    tabela_cat_rows, tabela_cont_rows, tempos_exec = [], [], []

    for idx_prop, prop in enumerate(proporcoes, start=1):
        display(HTML(f'<div class="header">🎯 PROPORÇÃO {idx_prop}/{len(proporcoes)}: {prop*100:.0f}% de missing</div>'))
        inicio = time.time()
        boot_out = bootstrap_misto(
            df_cod, prop, col_continuas, col_ordinais, col_nominais,
            indices_continuas, indices_ordinais,
            B=BOOTSTRAP_ITERACOES, m=M_IMPUTATIONS, random_seed=RANDOM_SEED,
            verbose_first=(prop == proporcoes[0]),
            use_conditional_mean=USE_CONDITIONAL_MEAN
        )
        fim = time.time()
        tempo_total = fim - inicio
        display(HTML(f'<div class="success-box">✅ Proporção {prop*100:.0f}% concluída em {tempo_total:.2f} segundos.</div>'))

        cat = boot_out['estat_cat']
        tabela_cat_rows.append({
            'Missing': f"{prop*100:.0f}%",
            'Acurácia': f"{cat['acuracia']['media']*100:.2f}%",
            'IC_Acurácia': f"[{cat['acuracia']['ic_inf']*100:.2f}%, {cat['acuracia']['ic_sup']*100:.2f}%]",
            'Precisão': f"{cat['precisao']['media']*100:.2f}%",
            'IC_Precisão': f"[{cat['precisao']['ic_inf']*100:.2f}%, {cat['precisao']['ic_sup']*100:.2f}%]",
            'Recall': f"{cat['recall']['media']*100:.2f}%",
            'IC_Recall': f"[{cat['recall']['ic_inf']*100:.2f}%, {cat['recall']['ic_sup']*100:.2f}%]",
            'F1': f"{cat['f1_score']['media']*100:.2f}%",
            'IC_F1': f"[{cat['f1_score']['ic_inf']*100:.2f}%, {cat['f1_score']['ic_sup']*100:.2f}%]",
            'Kappa': f"{cat['kappa']['media']:.3f}",
            'IC_Kappa': f"[{cat['kappa']['ic_inf']:.3f}, {cat['kappa']['ic_sup']:.3f}]",
            'MCC': f"{cat['mcc']['media']:.3f}",
            'IC_MCC': f"[{cat['mcc']['ic_inf']:.3f}, {cat['mcc']['ic_sup']:.3f}]",
            'Erro_Correlação': f"{cat['erro_correlacao']['media']:.4f}",
            'IC_ErroCor': f"[{cat['erro_correlacao']['ic_inf']:.4f}, {cat['erro_correlacao']['ic_sup']:.4f}]",
            'SMAE_Categórico': f"{cat['smae_cat']['media']:.3f}",
            'IC_SMAE_Cat': f"[{cat['smae_cat']['ic_inf']:.3f}, {cat['smae_cat']['ic_sup']:.3f}]",
            'SMAE_Contínuo': f"{cat['smae_cont']['media']:.3f}",
            'IC_SMAE_Cont': f"[{cat['smae_cont']['ic_inf']:.3f}, {cat['smae_cont']['ic_sup']:.3f}]",
            'Gower_Distance': f"{cat['gower_distance']['media']:.4f}",
            'IC_Gower': f"[{cat['gower_distance']['ic_inf']:.4f}, {cat['gower_distance']['ic_sup']:.4f}]"
        })

        cont = boot_out['estat_cont']
        tabela_cont_rows.append({
            'Missing': f"{prop*100:.0f}%",
            'RMSE': f"{cont['rmse_medio']['media']:.3f}",
            'IC_RMSE': f"[{cont['rmse_medio']['ic_inf']:.3f}, {cont['rmse_medio']['ic_sup']:.3f}]",
            'MAE': f"{cont['mae_medio']['media']:.3f}",
            'IC_MAE': f"[{cont['mae_medio']['ic_inf']:.3f}, {cont['mae_medio']['ic_sup']:.3f}]",
            'NMAE': f"{cont['nmae_medio']['media']:.4f}",
            'IC_NMAE': f"[{cont['nmae_medio']['ic_inf']:.4f}, {cont['nmae_medio']['ic_sup']:.4f}]",
            'MSE': f"{cont['mse_medio']['media']:.3f}",
            'IC_MSE': f"[{cont['mse_medio']['ic_inf']:.3f}, {cont['mse_medio']['ic_sup']:.3f}]",
            'MAPE(%)': f"{cont['mape_medio']['media']:.2f}",
            'IC_MAPE': f"[{cont['mape_medio']['ic_inf']:.2f}, {cont['mape_medio']['ic_sup']:.2f}]",
            'sMAPE(%)': f"{cont['smape_medio']['media']:.2f}",
            'IC_sMAPE': f"[{cont['smape_medio']['ic_inf']:.2f}, {cont['smape_medio']['ic_sup']:.2f}]",
            'Correlação': f"{cont['cor_medio']['media']:.3f}",
            'IC_Cor': f"[{cont['cor_medio']['ic_inf']:.3f}, {cont['cor_medio']['ic_sup']:.3f}]",
            'NRMSE': f"{cont['nrmse_medio']['media']:.4f}",
            'IC_NRMSE': f"[{cont['nrmse_medio']['ic_inf']:.4f}, {cont['nrmse_medio']['ic_sup']:.4f}]",
            'R²': f"{cont['r2_medio']['media']:.4f}",
            'IC_R²': f"[{cont['r2_medio']['ic_inf']:.4f}, {cont['r2_medio']['ic_sup']:.4f}]"
        })

        tempos_exec.append({
            'Proporção': f"{prop*100:.0f}%",
            'Tempo_Total': f"{tempo_total:.2f}",
            'Tempo_Medio_Boot': f"{boot_out['tempo_medio']:.2f}"
        })

    display(HTML('<div class="header">📊 RESULTADOS FINAIS - VARIÁVEIS ORDINAIS (size, day, time, smoker, sex)</div>'))
    df_cat_final = pd.DataFrame(tabela_cat_rows)
    display(df_cat_final.style.set_table_attributes('class="dataframe-table"'))

    display(HTML('<div class="header">📊 RESULTADOS FINAIS - VARIÁVEIS CONTÍNUAS (total_bill, tip)</div>'))
    df_cont_final = pd.DataFrame(tabela_cont_rows)
    display(df_cont_final.style.set_table_attributes('class="dataframe-table"'))

    display(HTML('<div class="header">⏱️ TEMPOS DE EXECUÇÃO POR PROPORÇÃO</div>'))
    df_tempo = pd.DataFrame(tempos_exec)
    display(df_tempo.style.set_table_attributes('class="dataframe-table"'))

    try:
        df_plot = pd.DataFrame(tabela_cat_rows)
        df_plot['prop'] = df_plot['Missing'].str.replace('%','').astype(float)/100
        df_plot['Acc'] = df_plot['Acurácia'].str.replace('%','').astype(float)/100
        plt.figure(figsize=(8,5))
        sns.lineplot(x='prop', y='Acc', data=df_plot, marker='o', color='blue')
        plt.xlabel("Proporção de missing")
        plt.ylabel("Acurácia")
        plt.title("Acurácia por proporção de dados faltantes - Dataset Tips (2 cont, 5 ord)")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()
    except Exception as e:
        display(HTML(f'<div class="warning-box">⚠️ Gráfico não pôde ser gerado: {e}</div>'))

    display(HTML('<div class="success-box">🎉 EXECUÇÃO CONCLUÍDA - TIPS COM 2 CONTÍNUAS E 5 ORDINAIS!</div>'))
```
